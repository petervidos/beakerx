```groovy
beakerx.foo = ['Alice', 'Bob', 'Ted']
beakerx.bar = [15, 32, 12]
beakerx.baz = [5, 3, 2]

beakerx.barchart = [
  x: 'Foo',
  y: 'Bar'      
]

beakerx.scatterplot = [
  x: 'Bar',
  y: 'Baz',
  color: 'Foo',
  geometry: 'circle'
]

OutputCell.HIDDEN
```

```groovy
%%python
from beakerx.object import beakerx
from ipyvizzu import Chart, Data, Config

data = Data()
data.add_series("Foo", beakerx.foo, type="dimension")
data.add_series("Bar", beakerx.bar, type="measure")
data.add_series("Baz", beakerx.baz, type="measure")

chart = Chart()
chart.animate(data)
chart.animate(Config(beakerx.barchart))
chart.animate(Config(beakerx.scatterplot))
```
